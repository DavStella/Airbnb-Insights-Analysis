```python
# Filter rows where price == 0 but accommodates is not 0
price_zero_accommodates_not_zero = paris_listings[(paris_listings["price"] == 0) & (paris_listings["accommodates"] != 0)]

# Display the filtered rows
print(price_zero_accommodates_not_zero)

```

```python
# Count the number of such rows
print(f"Number of rows with price == 0 and accommodates != 0: {len(price_zero_accommodates_not_zero)}")
```

# Observation
Observation: There are 62 rows with `price = 0` and 54 rows with `accommodates = 0`. However, there are 8 rows where `price = 0` but `accommodates` is greater than 0. This discrepancy suggests the possibility of:
1. **Potential Errors**: Listings with a price of 0 and accommodations available may indicate data entry mistakes or missing data.
   
2. **Promotional or Free Listings**: These could represent rare cases where hosts offer stays for free, such as for charity or promotions.
To ensure data integrity while being cautious, rows where both `price` and `accommodates` are 0 can be safely removed, while retaining the 8 rows with `price = 0` but non-zero accommodations, as they might represent valid cases of free or promotional listings.

```python
# Remove rows where both price == 0 and accommodates == 0
paris_listings = paris_listings[~((paris_listings["price"] == 0) & (paris_listings["accommodates"] == 0))]

# View the updated DataFrame
paris_listings.info()  # Check structure and size of the updated DataFrame

```

#### 1. Calculate the average price by neighborhood

```python

avg_price_by_neighborhood = paris_listings.groupby("neighbourhood")["price"].mean().sort_values()

# Convert the results to a DataFrame
avg_price_table = avg_price_by_neighborhood.reset_index()
avg_price_table.columns = ["Neighborhood", "Average Price (€)"]

# Display the table
print(avg_price_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average price by neighborhood
avg_price_by_neighborhood = paris_listings.groupby("neighbourhood")["price"].mean().sort_values()

# Create the horizontal bar chart
plt.figure(figsize=(10, 8))  # Adjust the figure size
sns.barplot(x=avg_price_by_neighborhood, y=avg_price_by_neighborhood.index, palette="viridis")

# Add title and labels
plt.title("Average Price by Neighborhood in Paris", fontsize=16)
plt.xlabel("Average Price (€)", fontsize=12)
plt.ylabel("Neighborhood", fontsize=12)

# Show the chart
plt.tight_layout()
plt.show()
```

#### Observation
Neighborhoods in Paris show a wide range of average prices, with Elysee (€211.25) being the most expensive and Menilmontant (€74.96) the most affordable, reflecting significant variability in listing costs.

##### 2. Showing the new host over time and average price

```python
paris_listings["host_year"] = paris_listings["host_since"].dt.year

# 1. Count of New Hosts Over Time
hosts_by_year = paris_listings.groupby("host_year").size()

plt.figure(figsize=(10, 6))
plt.plot(hosts_by_year.index, hosts_by_year.values, marker='o', color='b', label="New Hosts Count")
plt.title("Count of New Hosts Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of New Hosts", fontsize=12)
plt.ylim(0)  # Set y-axis limit to start from 0
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 2. Average Price Over Time
price_by_year = paris_listings.groupby("host_year")["price"].mean()

plt.figure(figsize=(10, 6))
plt.plot(price_by_year.index, price_by_year.values, marker='o', color='g', label="Average Price (€)")
plt.title("Average Price Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Price (€)", fontsize=12)
plt.ylim(0)  # Set y-axis limit to start from 0
plt.grid(True)
plt.legend()
plt.tight_layout
```

```python
# Create a new column for the year each host started
paris_listings["host_year"] = paris_listings["host_since"].dt.year

# Group by year to calculate the count of new hosts and average price
hosts_and_price_by_year = paris_listings.groupby("host_year").agg(
    new_hosts=("host_year", "size"),  # Count of new hosts
    average_price=("price", "mean")  # Average price
).reset_index()

# Display the resulting table
print(hosts_and_price_by_year)
```

```python
# Create a new column for the year each host started
paris_listings["host_year"] = paris_listings["host_since"].dt.year

# Data preparation
hosts_by_year = paris_listings.groupby("host_year").size()  # Count of new hosts
price_by_year = paris_listings.groupby("host_year")["price"].mean()  # Average price

# Create the figure and twin axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot new hosts count on the primary y-axis
ax1.plot(hosts_by_year.index, hosts_by_year.values, marker='o', color='b', label="New Hosts Count")
ax1.set_xlabel("Year", fontsize=12)
ax1.set_ylabel("Number of New Hosts", color='b', fontsize=12)
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_ylim(0)  # Set y-axis limit to start from 0

# Create a secondary y-axis for the average price
ax2 = ax1.twinx()
ax2.plot(price_by_year.index, price_by_year.values, marker='o', color='g', label="Average Price (€)")
ax2.set_ylabel("Average Price (€)", color='g', fontsize=12)
ax2.tick_params(axis='y', labelcolor='g')
ax2.set_ylim(0)  # Set y-axis limit to start from 0

# Add a title and grid
plt.title("New Hosts Count and Average Price Over Time", fontsize=16)
fig.tight_layout()
plt.grid(True)
plt.show()
```

### Observation
The number of new hosts increased significantly from 2008 to 2015, peaking at 12,147, while the average price has fluctuated, reaching a high of €159.64 in 2009 and dropping to €93.49 in 2021.
